```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
traindf = pd.read_csv("train.csv")
traindf.head()
```

```python
trainclean = traindf.loc[:,['id','Gender','Age','Height','Weight','family_history_with_overweight','FAVC','FCVC','NCP','SMOKE','CH2O','SCC','FAF','TUE','NObeyesdad']]
trainclean.head()
```

```python
columns_to_encode = ['Gender','family_history_with_overweight','FAVC','SMOKE','SCC','NObeyesdad']
labelencoder1 = LabelEncoder()
labelencoder2 = LabelEncoder()
labelencoder3 = LabelEncoder()
labelencoder4 = LabelEncoder()
labelencoder5 = LabelEncoder()
labelencoder6 = LabelEncoder()

```

```python
trainclean['Gender'] = labelencoder1.fit_transform(trainclean['Gender'])
trainclean['family_history_with_overweight'] = labelencoder2.fit_transform(trainclean['family_history_with_overweight'])
trainclean['FAVC'] = labelencoder3.fit_transform(trainclean['FAVC'])
trainclean['SMOKE'] = labelencoder4.fit_transform(trainclean['SMOKE'])
trainclean['SCC'] = labelencoder5.fit_transform(trainclean['SCC'])
trainclean['NObeyesdad'] = labelencoder6.fit_transform(trainclean['NObeyesdad'])

```

```python
trainclean.head()
```

```python
model = LogisticRegression(multi_class= 'multinomial',max_iter = 10000000)
```

```python
Xtrain = trainclean.iloc[:,1:14]
Xtrain.head()
```

```python
Ytrain = trainclean.iloc[:,14]
Ytrain.head()
```

```python
model.fit(Xtrain,Ytrain)
```

```python
testdf = pd.read_csv('test.csv')
```

```python
cleantest =  testdf.loc[:,['id','Gender','Age','Height','Weight','family_history_with_overweight','FAVC','FCVC','NCP','SMOKE','CH2O','SCC','FAF','TUE']]
cleantest.head()
```

```python
cleantest['Gender'] = labelencoder1.transform(cleantest['Gender'])
cleantest['family_history_with_overweight'] = labelencoder2.transform(cleantest['family_history_with_overweight'])
cleantest['FAVC'] = labelencoder3.transform(cleantest['FAVC'])
cleantest['SMOKE'] = labelencoder4.transform(cleantest['SMOKE'])
cleantest['SCC'] = labelencoder5.transform(cleantest['SCC'])


```

```python
cleantest.head()
```

```python
Xtest = cleantest.iloc[:,1:14]
Xtest.head()
```

```python
result =  model.predict(Xtest)
```

```python
resultdf = pd.DataFrame(result)
resultdf.head()
```

```python
resultdf['id'] = cleantest['id']
resultdf.columns = ['NObeyesdad','id']
```

```python
# Supondo que 'trainclean' é o seu DataFrame
cols = list(resultdf.columns)

# Trocando o lugar da primeira e segunda coluna
cols[0], cols[1] = cols[1], cols[0]

# Aplicando a nova ordem das colunas ao DataFrame
resultdf = resultdf[cols]
```

```python
resultdf['NObeyesdad'] = labelencoder6.inverse_transform(resultdf['NObeyesdad'])
```
