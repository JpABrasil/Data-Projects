```python
#Importing modules
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
#Importing Train Dataset
base = pd.read_csv("train.csv")
base.head()
```

```python
#Dropping some not useful features
droped = base.iloc[:,3:14]
droped.head()
```

```python
#Exploring Gender
droped['Gender'].value_counts()
```

```python
#Exploring Geography
droped['Geography'].value_counts()
```

```python
#Label enconding for Gender
droped['Gender'] = droped['Gender'].map({'Male':1,'Female':0})
droped.head()
```

```python
#Label enconding for Geography
droped['Geography'] = droped['Geography'].map({'France':0,'Spain':1, 'Germany':2})
droped.head()
```

```python
#Creating X variable
X = droped.iloc[:,0:10].values
print(X)
```

```python
#Creating Y variable
Y = droped.iloc[:,10].values
print(Y)
```

```python
#Creating and Fitting the model
model = LogisticRegression()
model.fit(X,Y)
```

```python
#Importing Test Dataset
test = pd.read_csv('test.csv')
test.head()
```

```python
#Treating the data
test['Gender'] = test['Gender'].map({'Male':1,'Female':0})
test['Geography'] = test['Geography'].map({'France':0,'Spain':1, 'Germany':2})
test.head()
```

```python
#Creating X_test variable
X_test =test.iloc[:,3:13].values
```

```python
#Prediction and Treating
prediction = model.predict(X_test)
base_prediction = np.column_stack((test['id'], prediction))
DataFrame_prediction = pd.DataFrame(base_prediction)
DataFrame_prediction.head()
```

```python
#Renaming Columns
DataFrame_prediction = DataFrame_prediction.rename(columns={0:'id',1:'Exited'})
DataFrame_prediction.head()
```

```python
#Exporting to csv
csv_result = DataFrame_prediction.to_csv('LogisticRegression Result.csv',index= False)
```
