```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression
```

```python
#Import Training Features
tra_f = pd.read_csv("dengue_features_train.csv")
tra_f.head()
#With this we can the DataSet have 2 categorical Features, but i judge them not effective for the accuracy of the model so i will drop them
```

```python
tra_f.info()
#With this we can the DataSet have 2 categorical Features, but i judge them not effective for the accuracy of the model so i will drop them
```

```python
#Importing Training Labels
tra_l= pd.read_csv("dengue_labels_train.csv")
tra_l.head()
```

```python
#Merge Them to Treat
tra_set =pd.merge(tra_f,tra_l, on=['city','year','weekofyear'],how = 'inner')
tra_set.head()
```

```python
#Treating features and Creating X and Y Variables
tra_set_droped = tra_set.drop(columns=['city','week_start_date','year','weekofyear'])
tra_set_noNaN = tra_set_droped.dropna()
X = tra_set_noNaN.iloc[:,0:20]
Y = tra_set_noNaN.iloc[:,20]
```

```python
#Creating the Model and Training it
model = LinearRegression()
model.fit(X,Y)
```

```python
#Import Test Features
test = pd.read_csv("dengue_features_test.csv")
```

```python
#Predicting
test_vals = test.drop(columns=['city','week_start_date','year','weekofyear']).dropna().values
predcition = model.predict(test_vals)
```

```python
#Agrouping Prediction
result = np.column_stack((test.dropna().values,predcition))
resultdf = pd.DataFrame(result)
resultdf = resultdf.iloc[:,list(range(0, 3)) + [24]]
resultdf = resultdf.rename(columns={0:'city',1:'year',2:'weekofyear',24:'total_cases'})
resultdf['total_cases'] = resultdf['total_cases'].astype(int)
resultdf.head()
```

```python
#Exporting
result_csv = resultdf.to_csv('LinearRegression Result.csv',index= False)
```
