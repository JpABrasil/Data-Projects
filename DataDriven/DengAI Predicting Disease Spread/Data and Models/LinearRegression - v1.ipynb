```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
#Import Training Features
trainfeatures = pd.read_csv("dengue_features_train.csv")
trainlabels = pd.read_csv("dengue_labels_train.csv")
train =pd.merge(trainfeatures,trainlabels, on=['city','year','weekofyear'],how = 'inner')
train.head()
```

```python
train.info()
#With this we can the DataSet have 2 categorical Features, but i judge them not effective for the accuracy of the model so i will drop them
```

```python
#Dropping Not Useful Features and Null rows
traindropna = train.dropna()
traindropfeatures = traindropna.drop(columns=['year','weekofyear','week_start_date'])
traindropfeatures.info()
```

```python
#Checking Categorical columns 
print(traindropfeatures['city'].unique())
#Its is needed to apply Label Encoding
labelencoder = LabelEncoder()
traindropfeatures['city'] = labelencoder.fit_transform(traindropfeatures['city'])

```

```python
#Treating features and Creating X and Y Variables
Xtrain = traindropfeatures.iloc[:,0:21]
Ytrain = traindropfeatures.iloc[:,21]
```

```python
#Creating the Model and Training it
model = LinearRegression()
model.fit(Xtrain,Ytrain)
```

```python
#Import Test Features
test = pd.read_csv("dengue_features_test.csv")
test.head()
```

```python
#Preprocessing
testdropna = test.dropna()
testdropfeatures = testdropna.drop(columns=['year','weekofyear','week_start_date'])

```

```python
#Applying Label encoder
testdropfeatures['city'] = labelencoder.transform(testdropfeatures['city'])
```

```python
#Making predictions
result = model.predict(testdropfeatures)
```

```python
#Checking Result
print(result)
#Rounding as int
resultint = np.round(result).astype(int)
print(resultint)
#Turning into Dataframe and indexing
resultdf = pd.DataFrame(resultint)
resultdf['year'] = testdropna['year']
resultdf['weekofyear'] = testdropna['weekofyear']
resultdf['week_start_date'] = testdropna['week_start_date']
resultdf.rename(columns={"0":"total_cases"})
```

```python
#Exporting
result_csv = resultdf.to_csv('LinearRegression Result.csv',index= False)
```
