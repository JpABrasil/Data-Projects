```python
# Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Lendo os dados e retirando valores nulos
stock_data = pd.read_csv("datasets/stock_data.csv", parse_dates=["Date"],index_col =["Date"]).dropna()
benchmark_data = pd.read_csv("datasets/benchmark_data.csv", parse_dates=["Date"],index_col =["Date"]).dropna()
```

```python
#Análise das ações em interesse
combined = pd.concat([stock_data,benchmark_data])
combined.describe()
stock_data.plot(subplots=True,title='Stock Data')
benchmark_data.plot(title='S&P 500 Data')
```
